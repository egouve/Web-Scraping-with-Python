```python
from bs4 import BeautifulSoup # Biblioteca de WebScrapping
```

# Raspagem em documento local

```python
with open ('dados/home.html', 'r') as html_file: #Abertura de documento local
    content = html_file.read()    
```

```python
soup = BeautifulSoup(content, 'lxml') #Permite que o bf4 use o lxml para ver o doc de forma mais legível
courses_html_tags = soup.find_all('h5') #find_all = encontrar todas as linhas que iniciem com h5
```

```python
for course in courses_html_tags:
    print(course.text) # Puxa apenas o texto dos h5
```

```python
course_cards = soup.find_all('div', class_='card') # Permite encontrar todas as linhas iniciadas com div, e com class = card
print(course_cards)                                                   # deve ser escrito "class_", por class é uma built-in function
```

```python
for course in course_cards:
    print(course.h5) # Puxa os h5 dentro dos course_cards
```

```python
for course in course_cards:
    course_name = course.h5.text # Puxa os texto dentro das h5 dos course_cards
    course_price = course.a.text.split()[-1] #o "a" é onde está armazenado o preço dos cursos, conforme documento
                                            #o split separa do texto o último elemento (-1)
    
    print(f'{course_name} costs {course_price}')
```

# Raspagem em sites

```python
import requests #Utilizado para fazer requests em páginas na internet
```

```python
#Puxando todos os trabalhos que exigem python no timejobs
#O .text indica que você está buscando apenas os textos na página. Sem isso, retornará 200, informando
#que o request foi um sucesso

html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=').text
soup = BeautifulSoup(html_text, 'lxml')
#print(html_text)
```

```python
job = soup.find('li', class_ = 'clearfix job-bx wht-shd-bx') #Traz apenas o primeiro resultado
print(job)
```

```python
# busca, dentro do job, o nome da empresa
company_name = job.find('h3', class_ = 'joblist-comp-name').text.replace(' ', '') # Remove os espaços em branco
print(company_name)
```

```python
skills = job.find('span', class_ = 'srp-skills').text.replace(' ', '')
print(skills)
```

```python
print(f'''
Company Name: {company_name}
Required Skills: {skills}
''')
```

```python
published_date = job.find('span', class_='sim-posted').span.text #busca o testo dentro do span que está dentro do span
print(published_date)
```

```python
jobs = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx') #Traz todos os resultados
```

```python
for job in jobs:
    company_name = job.find('h3', class_='joblist-comp-name').text.replace(' ', '')
    skills = job.find('span', class_ = 'srp-skills').text.replace(' ', '')
    published_date = job.find('span', class_='sim-posted').span.text

    print(f'''
Company Name: {company_name}
Required Skills: {skills}
Published Date: {published_date}
''')

```

```python
for job in jobs:
    published_date = job.find('span', class_='sim-posted').span.text 
    if 'few' in published_date: #Cria uma condição para puxar apenas jobs postados há poucos dias
        
        company_name = job.find('h3', class_='joblist-comp-name').text.replace(' ', '')
        skills = job.find('span', class_ = 'srp-skills').text.replace(' ', '')
    

        print(f'Company Name: {company_name}')
        print(f'Required Skills: {skills}')
        print(f'Published Date: {published_date}')

```

```python
#Transformando o programa numa função

def find_jobs():
    for job in jobs:
        published_date = job.find('span', class_='sim-posted').span.text 
        if 'few' in published_date: #Cria uma condição para puxar apenas jobs postados há poucos dias
        
            company_name = job.find('h3', class_='joblist-comp-name').text.replace(' ', '')
            skills = job.find('span', class_ = 'srp-skills').text.replace(' ', '')
            more_info = job.header.h2.a['href'] #puxa o href dentro do a dentro do h2 dentro do header do job
        
            if unfamiliar_skill not in skills:
                print(f'Company Name: {company_name.strip()}')
                print(f'Required Skills: {skills.strip()}')
                print(f'Published Date: {published_date.strip()}')
                print(f'More info: {more_info}')
                print('')
```

```python
import time
```

```python
#Criando uma automação para buscar os jobs a cada 600s (10min)

if __name__ == '__main__':
    while True:
        find_jobs()
        time_wait = 10
        print(f'Waiting {time_wait} minutes...')
        time.sleep(time_wait * 60)
```
